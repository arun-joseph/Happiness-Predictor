# Data Preprocessing

```python
import numpy as np
import pandas as pd
import csv
```

```python
#Processing 2015 data
with open('2015.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2015 = list(spamreader)
    data_2015 = np.asarray(data_2015)
    df_2015 = pd.DataFrame(data=data_2015[1:], columns=data_2015[0])
    df_2015.drop(columns=['Region', 'Happiness Rank', 'Standard Error'], axis=1, inplace=True)
    df_2015['Year'] = 2015
    df_2015 = df_2015[['Country', 'Year', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']]
    df_2015.columns = ['Country', 'Year', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']
    df_2015[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']] = df_2015[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']].astype(float)
```

```python
df_2015
```

```python
#Processing 2016 data
with open('2016.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2016 = list(spamreader)
    data_2016 = np.asarray(data_2016)
    df_2016 = pd.DataFrame(data=data_2016[1:], columns=data_2016[0])
    df_2016.drop(columns=['Region', 'Happiness Rank', 'Lower Confidence Interval', 'Upper Confidence Interval'], axis=1, inplace=True)
    df_2016['Year'] = 2016
    df_2016 = df_2016[['Country', 'Year', 'Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']]
    df_2016.columns = ['Country', 'Year', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']
    df_2016[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']] = df_2016[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']].astype(float)
```

```python
df_2016
```

```python
#Processing 2017 data
with open('2017.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data_2017 = list(spamreader)
    data_2017 = np.asarray(data_2017)
    df_2017 = pd.DataFrame(data=data_2017[1:], columns=data_2017[0])
    df_2017.drop(columns=['Happiness.Rank', 'Whisker.high', 'Whisker.low'], axis=1, inplace=True)
    df_2017['Year'] = 2017
    df_2017 = df_2017[['Country', 'Year', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.', 'Generosity', 'Dystopia.Residual']]
    df_2017.columns = ['Country', 'Year', 'Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']
    df_2017[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']] = df_2017[['Score', 'Economy', 'Family', 'Health', 'Freedom', 'Trust', 'Generosity', 'Dystopia']].astype(float)
```

```python
df_2017
```

```python
#Merging 2015, 2016, 2017
df_score = pd.concat([df_2015, df_2016, df_2017])
df_score.sort_values(by=['Country', 'Year'], inplace=True)
df_count = df_score.groupby('Country').count()
df_score.reset_index(inplace=True, drop=True)
```

```python
for country,row in df_count.iterrows():
    if(row['Year'] != 3):
        df_score.drop(df_score.loc[df_score['Country'] == country].index, inplace=True)
df_score.reset_index(inplace=True, drop=True)
df_score
```

```python
#Processing indicators
with open('fd1f9f89-994a-4025-b070-d32551a7c283_Data.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    data = list(spamreader)
    data = np.asarray(data)
    df = pd.DataFrame(data=data[1:-1556], columns=data[0])
    df.drop(columns=['Country Code', 'Series Name'], axis=1, inplace=True)
    df.columns = ['Country', 'Series', '2013', '2014', '2015', '2016', '2017', '2018']
    df.loc[df['2015'] == '..', '2015'] = {}
    df.loc[df['2016'] == '..', '2016'] = {}
    df.loc[df['2017'] == '..', '2017'] = {}
    df['2015'] = df['2015'].astype(float)
    df['2016'] = df['2016'].astype(float)
    df['2017'] = df['2017'].astype(float)
    df = df[['Country', 'Series', '2015', '2016', '2017']]
```

```python
df
```

```python
#Removing missing countries
countries = df['Country'].unique()
df_score.drop(df_score.loc[~df_score['Country'].isin(countries)].index, inplace=True)
df_score.reset_index(inplace=True, drop=True)
df_score
```

```python
#Combining the datasets
countries = df_score['Country'].unique()
for c in countries:
    indices = df.loc[df['Country'] == c].index
    if(indices.size == 0):
        continue
    
    c_ind = df_score.loc[df_score['Country'] == c].index
    for i in indices:
        df_score.at[c_ind[0], 'Ind '+str(i%33+1)] = df.loc[i]['2015']
        df_score.at[c_ind[1], 'Ind '+str(i%33+1)] = df.loc[i]['2016']
        df_score.at[c_ind[2], 'Ind '+str(i%33+1)] = df.loc[i]['2017']
df_score
```

```python
#Data interpolation
df_score.interpolate(inplace=True)
df_score = df_score.reindex(index=df_score.index[::-1])
df_score.interpolate(inplace=True)
df = df_score.reindex(index=df_score.index[::-1])
df
```

```python
#Saving as pickle file
df.to_pickle('happiness_data.pkl')
```

```python
#Saving as CSV file
df.to_csv('happiness_data.csv', sep=',', encoding='utf-8')
```
